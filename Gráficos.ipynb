```python
import pandas as pd
import numpy as sns 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
cafe = pd.read_csv("/content/predictmeancafe.csv",index_col=0)
cafe['Cultivo'] = "Cafe"
cacao = pd.read_csv("/content/predictmeancacao.csv",index_col=0)
cacao['Cultivo'] = "Cacao"
Input=cafe.append(cacao)
```

```python
f, axe = plt.subplots(1,1, figsize=(18,6))
Input.sort_values(by=['Cross-Validation'], ascending=False, inplace=True)
ax= sns.barplot(x='Cross-Validation', y='Model',hue='Cultivo', errwidth=0,data = Input, ax = axe)
axe.set_xlabel('Cross-Validaton Score', size=16)
axe.set_ylabel('Model')
axe.set_xlim(0,0.27)
for i in ax.containers:
    ax.bar_label(i,)
plt.show()
```
